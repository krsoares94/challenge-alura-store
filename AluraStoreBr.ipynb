### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Coloca todas as lojas em uma lista e cria uma lista com nomes.
lista_lojas = [loja,loja2,loja3,loja4]
nomes_lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]

# Calcula o faturamento das lojas
faturamento = []
for loja in lista_lojas:
  faturamento.append(float(loja["Preço"].sum()))

# Calcula o faturamento total e médio
faturamento_total = sum(faturamento)
faturamento_medio = faturamento_total / len(lista_lojas)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

plt.figure(figsize=(10, 6))

plt.bar(nomes_lojas, faturamento, color=['#5DA5DA', '#FAA43A', '#60BD68', '#F17CB0'])
plt.title('Faturamento por loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')

# Configuração do formato em reais
def formato_real(x, pos):
    return f'R$ {x:,.2f}'
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(formato_real))

# Adiciona os valores no topo de cada barra
for i, valor in enumerate(faturamento):
    plt.text(i, valor + 10000, f'R$ {valor:,.2f}',
             ha='center', va='bottom', fontsize=9, rotation=0)

# Adiciona uma linha horizontal indicando o faturamento médio
plt.axhline(y=faturamento_medio, color='r', linestyle='--', alpha=0.7)
plt.text(len(nomes_lojas)-0.5, faturamento_medio, f'Média: R$ {faturamento_medio:,.2f}',
         ha='right', va='bottom', color='red')

# Adiciona texto com o faturamento total na legenda
plt.figtext(0.5, 0.01, f'Faturamento Total: R$ {faturamento_total:,.2f}',
            ha='center', fontsize=12,)

plt.ylim(1000000, 1600000)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('faturamento_loja.png', dpi=300, bbox_inches='tight')
plt.subplots_adjust(bottom=0.18)  # Aumenta o espaço na parte inferior para o texto
plt.show()
```

# 2. Vendas por Categoria


```python
# Calcula a quantidade de produtos vendidos por categoria em cada loja
categoria_loja1 = loja.groupby("Categoria do Produto").size()
categoria_loja2 = loja2.groupby("Categoria do Produto").size()
categoria_loja3 = loja3.groupby("Categoria do Produto").size()
categoria_loja4 = loja4.groupby("Categoria do Produto").size()

# Armazena os resultados em um dicionário para fácil acesso
resultados_por_loja = {
    "Loja 1": categoria_loja1,
    "Loja 2": categoria_loja2,
    "Loja 3": categoria_loja3,
    "Loja 4": categoria_loja4
}

# Encontra a categoria mais vendida no geral
todas_categorias = pd.concat([categoria_loja1, categoria_loja2, categoria_loja3, categoria_loja4])
categoria_mais_vendida = todas_categorias.groupby(todas_categorias.index).sum().idxmax()
categoria_quantidade_mais_vendida = todas_categorias.groupby(todas_categorias.index).sum().max()
```

```python
# Importando as bibliotecas necessárias
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# DataFrame combinado a partir do dicionário de resultados e preparando os dados
df_categorias = pd.DataFrame(resultados_por_loja)
df_plot = df_categorias.T

# Definindo um esquema de cores para as categorias
categorias = df_plot.columns
num_categorias = len(categorias)
cores = plt.cm.tab20(np.linspace(0, 1, num_categorias))

plt.figure(figsize=(10, 6))
y_pos = np.arange(len(df_plot.index))
largura_barra = 0.8  # Largura total disponível para todas as categorias em cada loja

# Calculando a largura para cada categoria
largura_categoria = largura_barra / num_categorias

# Plotando barras para cada categoria
for i, categoria in enumerate(categorias):
    # Deslocamento para posicionar as barras de cada categoria lado a lado
    deslocamento = i * largura_categoria - largura_barra/2 + largura_categoria/2

    plt.barh(y_pos + deslocamento, df_plot[categoria],
             height=largura_categoria,
             color=cores[i],
             label=categoria)

plt.yticks(y_pos, df_plot.index)


plt.title('Quantidade de Produtos Vendidos por Categoria em Cada Loja', fontsize=14)
plt.xlabel('Quantidade de Produtos', fontsize=12)
plt.ylabel('Lojas', fontsize=12)
plt.legend(title='Categorias', bbox_to_anchor=(1.0, 1), loc='upper left')

# Adicionando anotação explicativa
plt.figtext(0.5, -0.01,
            f"A categoria mais vendida no geral é '{categoria_mais_vendida}' com {categoria_quantidade_mais_vendida} unidades vendidas.",
            ha='center', fontsize=12)

plt.savefig('quantidade_por_categoria.png', dpi=300, bbox_inches='tight')
plt.subplots_adjust(bottom=0.5)  # Aumenta o espaço na parte inferior para o texto
plt.grid(axis='x', color="black", linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import textwrap
import squarify

# Criando treemaps para todas as lojas com cores distintas e sem padding
plt.figure(figsize=(12, 8))

# Primeiro, identificamos todas as categorias únicas em todas as lojas
todas_categorias = set()
for dados_loja in resultados_por_loja.values():
    todas_categorias.update(dados_loja.index)

# Criamos um dicionário para mapear cada categoria a uma cor específica
# Isso garante que a mesma categoria tenha a mesma cor em todos os treemaps
import matplotlib.colors as mcolors

# Obtendo uma lista de cores distintas
cores_base = list(mcolors.TABLEAU_COLORS.values())  # 10 cores distintas do Tableau
cores_base.extend(list(mcolors.CSS4_COLORS.values())[:40])  # Adicionando mais cores do CSS4

# Mapeando categorias para cores
mapa_cores = {}
for i, categoria in enumerate(todas_categorias):
    mapa_cores[categoria] = cores_base[i % len(cores_base)]

# Configurando os subplots
for i, (nome_loja, dados_loja) in enumerate(resultados_por_loja.items(), 1):
    plt.subplot(2, 2, i)

    # Extraindo os dados para esta loja
    valores = dados_loja.values
    categorias_loja = dados_loja.index
    labels = [textwrap.fill(f"{cat}: {val}", width=20).upper() for cat, val in zip(categorias_loja, valores)]


    # Obtendo as cores para essas categorias específicas
    cores = [mapa_cores[cat] for cat in categorias_loja]

    # Criando o treemap
    squarify.plot(sizes=valores,
                  label=labels,
                  color=cores,
                  text_kwargs = {'fontsize': 8, 'weight': '600', 'color': 'white'},
                  alpha=1.0,  # Sem transparência
                  pad=0)      # Sem padding

    plt.title(f'Produtos Vendidos por Categoria - {nome_loja}', fontsize=14)
    plt.axis('off')

# Adicionando anotação explicativa
plt.figtext(0.5, 0.01,
            'O tamanho de cada retângulo representa a quantidade total de produtos vendidos em cada categoria',
            ha='center', fontsize=12)

plt.tight_layout()
plt.savefig('treemap_categoria_produto.png', dpi=300, bbox_inches='tight')
plt.subplots_adjust(bottom=0.05)  # Aumenta o espaço na parte inferior para o texto
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
import matplotlib.pyplot as plt
import numpy as np

# Calcular média de avaliação em cada loja e armazenando para fácil acesso
media_loja1 = loja["Avaliação da compra"].mean()
media_loja3 = loja3["Avaliação da compra"].mean()
media_loja2 = loja2["Avaliação da compra"].mean()
media_loja4 = loja4["Avaliação da compra"].mean()
medias = [media_loja1, media_loja2, media_loja3, media_loja4]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Criar figura
plt.figure(figsize=(8, 4))
plt.style.use('ggplot')

# Criar gráfico de barras
bars = plt.bar(nomes_lojas, medias, color = ['#5DA5DA', '#FAA43A', '#60BD68', '#F17CB0'], width=0.6)

# Adicionar rótulos nas barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.05,
             f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

# Títulos e labels
plt.title('Média de Avaliações por Loja', fontsize=16, fontweight='bold')
plt.ylabel('Média de Avaliação', fontsize=12)
plt.ylim(0, 5.0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar layout e salvar
plt.tight_layout()
plt.ylim(3.0, 5)
plt.savefig('comparacao_avaliacoes_lojas.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Calcula quantidade de cada produto vendido em cada loja
produto_loja1 = loja.groupby("Produto").size().sort_values(ascending=False)
produto_loja2 = loja2.groupby("Produto").size().sort_values(ascending=False)
produto_loja3 = loja3.groupby("Produto").size().sort_values(ascending=False)
produto_loja4 = loja4.groupby("Produto").size().sort_values(ascending=False)

# Armazena resultados em um dicionário para fácil acesso
resultado_produto_loja = {
    "Loja 1": produto_loja1,
    "Loja 2": produto_loja2,
    "Loja 3": produto_loja3,
    "Loja 4": produto_loja4
}

# Encontrar produto mais vendido no geral
todos_produtos = pd.concat([produto_loja1, produto_loja2, produto_loja3, produto_loja4])
produto_mais_vendido = todos_produtos.groupby(todos_produtos.index).sum().idxmax()
produto_menos_vendido = todos_produtos.groupby(todos_produtos.index).sum().idxmin()
quantidade_mais_vendida = todos_produtos.groupby(todos_produtos.index).sum().max()
quantidade_menos_vendida = todos_produtos.groupby(todos_produtos.index).sum().min()
```

```python
import matplotlib.pyplot as plt

# Criar listas para armazenar os produtos e quantidades por loja
produtos_mais = []
quantidades_mais = []
produtos_menos = []
quantidades_menos = []

# Preencher as listas com os dados do dicionário
for nome_loja, dados in resultado_produto_loja.items():
    produtos_mais.append(dados.idxmax())
    quantidades_mais.append(dados.max())
    produtos_menos.append(dados.idxmin())
    quantidades_menos.append(dados.min())

# Criar gráfico
x = np.arange(len(nomes_lojas))  # posição das lojas no eixo x
largura = 0.35

plt.figure(figsize=(12, 7))
plt.style.use('ggplot')

# Barras para produtos mais vendidos
barras_mais = plt.bar(x - largura/2, quantidades_mais, width=largura, label='Mais Vendido', color='#5DA5DA')

# Barras para produtos menos vendidos
barras_menos = plt.bar(x + largura/2, quantidades_menos, width=largura, label='Menos Vendido', color='#F15854')

# Adiciona os nomes dos produtos nas barras
for i in range(len(nomes_lojas)):
    plt.text(x[i] - largura/2, quantidades_mais[i] + 0.1, produtos_mais[i], ha='center', va='bottom', fontsize=9, rotation=0)
    plt.text(x[i] + largura/2, quantidades_menos[i] + 0.1, produtos_menos[i], ha='center', va='bottom', fontsize=9, rotation=0)

# Eixos e legendas
plt.xticks(x, nomes_lojas)
plt.ylabel('Quantidade Vendida')
plt.title('Produto Mais e Menos Vendido por Loja', fontweight='bold')
plt.legend()
plt.tight_layout()

# Adicionando anotação explicativa
plt.figtext(0.5, 0.01,
            f"O produto mais vendido no geral é '{produto_mais_vendido}' com {quantidade_mais_vendida} unidades vendidas.\n O produto menos vendido no geral é '{produto_menos_vendido}' com {quantidade_menos_vendida} unidades vendidas.",
            ha='center', fontsize=12)

plt.savefig('mais_menos_vendido_por_loja.png', dpi=300, bbox_inches='tight')
plt.subplots_adjust(bottom=0.1)  # Aumenta o espaço na parte inferior para o texto
plt.ylim(20,)
plt.show()
```

# 5. Frete Médio por Loja


```python
# Cria listas para armazenar os valores
media_fretes = []
total_fretes = []

# Calcula os valores para cada loja
for df in lista_lojas:
    media_fretes.append(df['Frete'].mean())
    total_fretes.append(df['Frete'].sum())

df_frete = pd.DataFrame({
    'Loja': nomes_lojas,
    'Média Frete': media_fretes,
    'Total Frete': total_fretes
})
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_frete['Total Frete'], df_frete['Média Frete'], color=['#5DA5DA', '#FAA43A', '#60BD68', '#F17CB0'], s=150)


for i in range(len(df_frete)):
    plt.text(df_frete['Total Frete'][i],
             df_frete['Média Frete'][i] +.2,
             df_frete['Loja'][i],
             fontsize=12,
             ha='right',
             va='center')

plt.xlabel('Total de Frete (R$)')
plt.ylabel('Média de Frete (R$)')
plt.title('Dispersão: Média x Total de Frete por Loja')
plt.grid(True)
plt.tight_layout()
plt.ylim(31, 35.5)
plt.xlim(72000,82000)
plt.show()
```
